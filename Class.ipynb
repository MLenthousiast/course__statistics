## Class 1. Introduction.
Statistics is about:
- Data collection.
- Data description.
- Inference. Making a prediction about a population using a (representative) sample.  

Parameter vs statistic:
- Parameter: numerical summary of the population.
- Statistic: numerical summary of the sample.

Variable types:
- Qualitative/categorical.
    - Nominal.
    - Ordinal (convertible to interval).
- Quantitative.
    - Interval. Discrete or continuous.


## Class 2. Data and probabilities.
Definitions.
- _Sample space_ (uitkomstenruimte) is the collection of all possible outcomes of an experiment.
- _Event_ (gebeurtenis) is a subset of the sample space.
- _Probability distribution_ (kansverdeling) assigns a value to each possible event.
- _Marginal probability_ (marginale kans) is the probability of an event occuring ($p(A)$).
- _Joint probability_ (gezamenlijke kans) is the probability of two events occuring at the same time ($p(A,B)$).
- _Conditional probability_ (conditionale kans) given event A, what is the probability for event B ($p(B\mid A)$)?
- _Independence_ (onafhankelijkheid) $p(B\mid A)=p(B)$, so $p(AB)=p(A)p(B)$.
- _Sampling error_ (steekproeffout): is the error of the sample statistic wiht regard to the population parameter.
- _Bias_ (sytematische fout): systematic sampling error. Nonresponse bias.
- _Big data_: $n = all$.

Frequentist vs Bayesian.
- _Frequentist_: probability tells us what happpens when a experiment is repeated many times.
- _Bayesian_: probability tells us what happens when a single experiment is performed.

Data collection.
- Sample survey.
- Experiment.

Randomization.
- _Simple random sample_ of size $n$: each possible sample of size $n$ as an equal probability of being selected. A sampling frame (list with all subjects of a population) is necessary.
- _Probability sampling_: stratified (fixes the underrepresentation of certain groups) and cluster random sampling (fixes feasibility, each cluster is representative of the population).
- _Nonprobability sampling_: volunteers for example.

Experimental vs observational.
- _Experimental_: adjust variables for different groups.
- _Observational_: observe.

## Class 4. Variables and probability distributions.
- _Stochastic variable_: assigns real value to each outcome ($X:S\rightarrow  	\mathbb{R})$

Probability distribution types.
- _Population distritubtion_: for example gaussian with $\sigma$ and $\mu$.
- _Sample (data) distribution_: sample mean $\bar{y}$ with standard deviation $s$.
- _Sampling distribution_: probability distribution for the possible values of a statistic.

Frequency vs probability distribution.
- _Frequency_: Assigns a frequency to a variable.
- _Probability_: Assings a probability to a variable.

Probability distributions.
- _Discrete_: assings probability to each discrete event.
- _Continuous_: assings probability to an interval in the form of Area Under Curve (AUC). Single events have a probability of $0$.

Probability distribution variants.
- _Binomial_: divides population into proportions.
- _Gauss (Normal)_. Standard normal has $\mu=0$ and $\sigma=1$.

Proportions.
- $\pi$ fall in group A, having value $1$. 
- $1-\pi$ fall in group NOT A, having value $0$.
- $\mu = n\pi$
- $\sigma = \sqrt{n\pi(1-\pi)}$

Standard deviation of a probability distribution.
- $\sigma = \sqrt{\sum_{i}{(x_{i}-\mu)^{2}p(x_{i})}}$. Note that each $x_{i}$ has its own probability, so no division by $n$.
- $z = \frac{y-\mu}{\sigma}$, or number of std.dev. from the mean.

Sampling distribution (Steekproevenverdeling): lists the possible values of a statistic.  
_Sample mean_: $\bar{y}$ has mean $\mu$ and standard error $\sigma_{\bar{y}}=\frac{\sigma}{\sqrt{n}}$.

Central Limit Theorem (CLT).
- For random sampling with $n \geq 30$ the sampling distribution of the sample mean is a normal distribution.

## Class 5. Probability distributions.
- _Probability mass function_ (PMF): assigns probability to discrete values.
- _Probability density function_ (PDF): "assigns" probability to continues variables.
- _Cumulative distribution funtion_ (CDF): gives the summed probability up to a certain point.
- _Point mass distribution_ (PMS) (Puntmassaverdeling): function $F(x)$ classifies points into $0$ and $1$.  
- _Discrete uniform distribution_: uniform.
- _Continuous uniform distribution_: uniform.
- _Bernoulli distribution_: $f(x)=p^{x}(1-p)^{1-x}$ for $x$ between $0$ and $1$.
- _Binomial distribution_: $\binom{n}{x}p^{x}(1-p)^{n-x} $ or $0$.
- _Normal distribution_: normal.
- Others: Poisson, t, chi-squared

## Class 6. Estimation.
Quantitative vs qualitative.
- Quantitative: inference about the mean. 
- Qualitative: inference about the proportion. Needs at least $15$ observations for each category.

Preferably unbiased and efficient (small error).
- _Point estimate_: single statistic value (in case of proportion $\hat{\pi}$).
- _Interval estimate_: a confidence interval (CI) around a certain point with a fixed confidence level of containing the statistic value.
- _Confidence interval_ (CI): interval of numbers believed to contain the parameter value with a certain confidence. Form: point estimate +- margin of error.
- _Confidence interval (proportion)_:   
**In theory:**  
Standard deviation: $\sigma_ = \sqrt{\pi(1-\pi)}{n}$   
Standard error of sample mean: $\sigma_{\hat{\pi}} = \frac{\sigma}{\sqrt{n}} = \sqrt{\frac{\pi(1-
\pi)}{n}}$ with $\pi$ within $\hat{\pi}\pm z(\sigma_{\hat{\pi}})$.  
**In practice:**  
$se = \sqrt{\frac{\hat{\pi}(1-\hat{\pi})}{n}}$ using $\hat{\pi}\pm z(se)$ with $p=\hat{\pi}$
- _Confidence interval (mean)_:  
**In theory:**  
$\sigma_{\bar{y}} = \frac{\sigma}{\sqrt{n}}$, using $\bar{y}\pm z(\sigma_{\bar{y}})$.  
**In practice:**  
$se = \frac{s}{\sqrt{n}}$, using $\bar{y}\pm z(se)$ with $s$ being the point estimate of $\sigma_{\bar{y}}$.

_T-distribution_:  
- Bell-like but a bit larger, depends on degrees of freedom ($df=n-1$). The larger $df$ the more Bell-like, nearly identical when $n\geq30$.
- Assume normal population distribution. Then a $95\%$ CI for $\mu$ is: $\bar{y}\pm t_{0.25}(se)$, with $se=\frac{s}{\sqrt{n}}$ and $df=n-1$.
- Somewhat robust to violation of normal population distribution assumption.


## Class 7. Statistical tests.
Significance tests/Hypothesis test.
1. Assumptions. Quantitative? Random sampling? Population distribution?
1. Hypothesis. $H_{0}$ and $H_{a}$.
1. Test statistic. Find number of standard errors between sample estimate and $H_{0}$.
1. P-value. Assuming null hypothesis to be true, what is the probability of observing a given sample? The smaller it is, the less confident we are in $H_{0}$.
1. Conclusion. Report P-value or set in advance a significance level ($\alpha$) with a certain value (typically 0.05 or 0.01).

Significance test for mean.
1. Assumptions.  
Randomization, quantitative variable, normal population distribution.
1. Hypothesis.  
$H_{0}:\mu=\mu_{0}$  
$H_{a}:\mu\neq\mu_{0}$ thus 2-sided.
1. Test statistic.  
$t=\frac{\bar{y}-\mu_{0}}{se}$ where $se = \frac{s}{\sqrt{n}}$ and $df=n-1$.
1. P-value.  
Given observed t-value and $df$ find the probability.
1. Conclusion.  
If greater then $0.05$ then reject $H_{0}$.

Note: When P-value $\leq0.05$ in two-sided, $95\%$ CI for $\mu$ does not contain $H_{0}$ value of $\mu$.

Significance test for proportion.
1. Assumptions. 
Categorical variable, randomization, large sample.
1. Hypothesis.  
$H_{0}:\pi=\pi_{0}$  
$H_{a}:\pi\neq\pi_{0}$ thus 2-sided.
1. Test statistic.  
$z = \frac{\hat{\pi}-\pi_{0}}{\sigma_{\hat{pi}}}=\frac{\hat{\pi}-\pi_{0}}{\sqrt{\pi_{0}(1-\pi_{0})/n}}$  
Note: $\sigma_{\hat{pi}}=se_{0}=\sqrt{\pi_{0}(1-\pi_{0})/n}$, not $se=\sqrt{\hat{\pi}(1- \hat{\pi} )/n}$ as in CI.
1. P-value.   
1. Conclusion.  
Reject $H_{0}$if P-value$\leq\alpha$.

Error types.  

|               | Test result                        |
|---------------|----------------|-------------------|
| **True state** | Reject $H_{0}$ | No Reject $H_{0}$ |
| True $H_{0}$  | Type 1 error   |                   |
| False $H_{0}$ |                | Type 2 error      |

Significance.
Statistical significance is not the same as practical significance.
Some test are just by chance significant.
